```python
import re

def extract_numbers(text):
    # 使用正则表达式匹配数字
    numbers = re.findall(r'\d+', text)
    # 将匹配到的数字连接起来并转换为整数
    return int(''.join(numbers)) if numbers else None

review_count['review_count'] = review_count['review_count'].apply(extract_numbers)
review_count['fake_review_count'] = review_count['fake_review_count'].apply(extract_numbers)
```
