# merge sampling and all restaurants

```python
all_restaurant = pd.read_csv('all_restaurant.csv')
```

```python
all_restaurant.shape
```

```python
table = all_restaurant.head()
display(table)
```

```python
merged_data = all_restaurant.merge(restaurants, on=['title', 'city'], how='right')
```

```python
table = merged_data.head(20)
display(table)
```

# drop_duplicates

```python
# 合并结果为新的数据集
merged_data = merged_data.drop_duplicates(subset=['title', 'city', 'link'], keep='first')
# 打印结果
print(merged_data)
```

```python
merged_data.shape
```

# 전처리 review

```python
merged_data['review'] = merged_data['review'].str.replace(r'\(|\)', '', regex=True)
```

```python
merged_data['review'] = merged_data['review'].str.rsplit(' ', n=1).str[0]
```

```python
merged_data['review_n'] = pd.to_numeric(merged_data['review'], errors='coerce')
```

```python
merged_data['review_k'] = merged_data['review'].where(merged_data['review'].str.contains('k'), None)
```

```python
merged_data['review_k']= merged_data['review_k'].str.replace('k', '').astype(float) * 1000
```

```python
merged_data['review_n'].fillna(merged_data['review_k'], inplace=True)
```

```python
merged_data.drop(columns=['review_k','review'], inplace=True)
```

```python
missing_review_n = merged_data[merged_data['review_n'].isna()]
print(missing_review_n)
```

```python
table = merged_data.head(10)
display(table)
```

# 전처리 missing data in 'point' and 'review'

```python
merged_data.loc[merged_data['title'] == 'Fat Belly Deli', 'point'] = '0'
merged_data.loc[merged_data['title'] == 'Fat Belly Deli', 'review_n'] = '0'

merged_data.loc[merged_data['title'] == 'Pf Burger', 'point'] = '5'
merged_data.loc[merged_data['title'] == 'Pf Burger', 'review_n'] = '1'

merged_data.loc[merged_data['title'] == 'Highlands Bar & Grill', 'review_n'] = '1'
```

```python
merged_data['point'] = merged_data['point'].astype(float)
merged_data['review_n'] = merged_data['review_n'].astype(float)
```

# 전처리 'text'

```python
import re
import html

def remove_non_ascii(text):
    # 先解码 HTML 实体编码，然后再去除非 ASCII 字符
    text = html.unescape(str(text))
    return re.sub(r'[^\x00-\x7F]+', '', text)

# 在应用函数之前，确保 'text' 列中的所有值都是字符串
merged_data['title'] = merged_data['title'].apply(remove_non_ascii)
```

# save

```python
merged_data.drop(columns=['link'], inplace=True)
```
